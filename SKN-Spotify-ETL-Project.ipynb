```python
#Grab a playlist URL from any given Spotify playlist -> (Global Top 50)
playlist_link = 'https://open.spotify.com/playlist/37i9dQZEVXbMDoHDwVN2tF'
```

```python
#Grab the URI specific to the playlist
playlist_URI = playlist_link.split("/")[-1]
```

```python
#Use the authorized spotify object to extract playlist information from the playlist_URI
data = sp.playlist_tracks(playlist_URI)
data
```

```python
data['items'][0]['track']['artists'][0]
```

```python
#Album Data:

album_list = []
print(data['items'][0].keys())
for row in data['items']:
    #print(type(row))
    album = row['track']['album']
    album_id = album['id']
    album_name = album['name']
    album_artists = album['artists'][0]['name']
    album_release_date = album['release_date']
    album_total_tracks = album['total_tracks']
    
    album_json = {'album_id':album_id, 
                  'album_name':album_name,
                  'album_artists':album_artists,
                  'album_release_date':album_release_date, 
                  'album_total_tracks':album_total_tracks}
    
    album_list.append(album_json)
```

```python
#Artists Data

artist_list = []
for row in data['items']:
    for key, value in row.items():
        if key == 'track':
            for artist in value['artists']:
                artist_id = artist['id']
                artist_name = artist['name']
                artist_url = artist['href']
                
                artist_json = {'artist_id':artist_id,
                               'artist_name':artist_name, 
                               'artist_url':artist_url}
                
                artist_list.append(artist_json)
         
```

```python
#Song Data

song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['href']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['artists'][0]['id']
    
    song_json = {'song_id':song_id, 
                 'song':song, 
                 'song_duration':song_duration, 
                 'song_url':song_url, 
                 'song_added':song_added, 
                 'album_id':album_id, 
                 'artist_id':artist_id}
    
    song_list.append(song_json)
```
